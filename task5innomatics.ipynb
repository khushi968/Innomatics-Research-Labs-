# social media


```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


analyze_engagement([200, 300, 150, 400])
```

# Healthcare

```python
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


check_medicine_stock(6)
```

# Agriculture

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)



rainfall_checker([70, 80, 65, 75], 70)
```

 # Social Media â€“ Duplicate Account Detection

```python
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

detect_duplicates(["john", "emma", "john", "alex"])
```

# System Monitoring

```python
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

check_system_health(7)
```

# Banking

```python
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


check_transaction(60000)
```

# E-Learning

```python
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


check_attendance([1,1,0,1,1,1,0,1])
```

# Smart Electricity Bill 

```python
def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


bill, status = calculate_electricity_bill(250)
print("Total Bill:", bill)
print("Usage Level:", status)
```
