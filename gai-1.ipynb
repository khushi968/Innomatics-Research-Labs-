## Task 1: Replace Hardcoded Prompts

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
simple_explainer = PromptTemplate(
    input_variables=["topic"],
    template="Describe {topic} in an easy way for a beginner."
)

print("Task 1 Result:")
print(simple_explainer.format(topic="Cloud Computing"))
```

## Task 2: Build a Prompt with Multiple Inputs

```python
learning_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to {audience} using a {tone} style."
)

examples = [
    {"topic": "Artificial Intelligence", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("Task 2 Results:\n")
for item in examples:
    print(learning_prompt.format(
        topic=item["topic"],
        audience=item["audience"],
        tone=item["tone"]
    ))
```

## Task 3: Create Prompt Variations

```python
teacher_mode = PromptTemplate(
    input_variables=["topic"],
    template="Teach {topic} in a simple step-by-step manner."
)

interview_mode = PromptTemplate(
    input_variables=["topic"],
    template="Generate 3 interview questions based on {topic}."
)

story_mode = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} like a short interesting story."
)

topic_name = "Machine Learning"

print("Task 3 Results:\n")
print("Teaching Style:")
print(teacher_mode.format(topic=topic_name))

print("\nInterview Style:")
print(interview_mode.format(topic=topic_name))

print("\nStorytelling Style:")
print(story_mode.format(topic=topic_name))
```

## Task 4: Build a Role-Based Chat Prompt System

```python
def build_chat_prompt(role, topic):
    role_map = {
        "teacher": "You are an experienced teacher who explains concepts clearly.",
        "interviewer": "You are a technical interviewer asking smart and relevant questions.",
        "motivator": "You are a positive mentor who inspires students to learn."
    }

    system_text = role_map.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_text),
        ("user", "Please help me understand {topic}.")
    ])

    return chat_template.format_messages(topic=topic)

messages = build_chat_prompt("interviewer", "Neural Networks")

print("Task 4 Result:\n")
for msg in messages:
    print(f"{msg.type.upper()}: {msg.content}")
```

## Task 5: Validate User Inputs

```python
def check_inputs(audience, tone):
    allowed_audience = ["beginner", "intermediate", "expert"]
    allowed_tone = ["formal", "casual", "fun"]

    if audience not in allowed_audience:
        raise ValueError("Audience must be beginner, intermediate, or expert.")

    if tone not in allowed_tone:
        raise ValueError("Tone must be formal, casual, or fun.")

    return "Inputs are valid."

print("Task 5 Result:")
print(check_inputs("beginner", "casual"))

try:
    print(check_inputs("students", "cool"))
except ValueError as e:
    print("Error:", e)
```

## Task 6: Build a Prompt Generator Function

```python
def create_prompt(topic, audience, tone, style):
    check_inputs(audience, tone)

    prompt_options = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Teach {topic} to a {audience} learner using a {tone} tone."
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Create interview questions on {topic} for a {audience} learner in a {tone} style."
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} to a {audience} learner in a {tone} storytelling format."
        )
    }

    if style not in prompt_options:
        raise ValueError("Style must be teaching, interview, or storytelling.")

    return prompt_options[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print("Task 6 Result:\n")
print(create_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Reuse One Template with Different Inputs

```python
universal_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} learners in a {tone} tone."
)

sample_inputs = [
    {"topic": "Cyber Security", "audience": "beginner", "tone": "formal"},
    {"topic": "Java", "audience": "intermediate", "tone": "casual"},
    {"topic": "Blockchain", "audience": "expert", "tone": "formal"},
    {"topic": "Data Analytics", "audience": "beginner", "tone": "fun"},
    {"topic": "Web Development", "audience": "intermediate", "tone": "casual"}
]

print("Task 7 Results:\n")
for data in sample_inputs:
    print(universal_template.format(
        topic=data["topic"],
        audience=data["audience"],
        tone=data["tone"]
    ))
```
