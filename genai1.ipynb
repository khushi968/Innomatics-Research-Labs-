### 1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing (NLP), "Love" and "love" are treated as different tokens if case normalization is not applied. This is because most NLP models are case-sensitive by default. As a result, treating them separately increases the vocabulary size unnecessarily and can reduce model efficiency. To avoid this, text is usually converted to lowercase so that both "Love" and "love" are treated as the same word.


### 2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset may contain a lot of unnecessary words such as "is", "the", "and", etc., which do not add much meaning to the text. This increases noise in the data, makes processing slower, and can negatively affect model performance. Removing stopwords helps the model focus on meaningful words.


### 3. Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords is not always beneficial. In some cases, it can change the meaning of a sentence:

**a) Sentiment Analysis:**
Example: "I am not happy"
If the word "not" is removed, the sentence becomes "I am happy", which completely changes the meaning.

**b) Question Answering / Chatbots:**
Example: "What is your name?"
If stopwords like "what" and "is" are removed, the sentence structure becomes unclear, making it difficult for systems to understand the intent.


### 4. Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both techniques used to reduce words to their base form, but they work differently:

* **Stemming:**
  It removes prefixes or suffixes from words using simple rules, which may result in non-meaningful words.
  Example: "running" → "runn"

* **Lemmatization:**
  It converts words to their base or dictionary form using vocabulary and morphological analysis.
  Example: "running" → "run"

Lemmatization is more accurate and meaningful, while stemming is faster but less precise.


```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # Lowercase
    text = text.lower()
    
    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    tokens = text.split()
    
    # Remove short tokens except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    clean_sentence = " ".join(tokens)
    
    return tokens, clean_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
```

```python
def analyze(tokens):
    if len(tokens) == 0:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total
    
    return total, unique, round(avg_len, 2)

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = analyze(tokens)
    
    print("\nSentence:", sentence)
    print("Total:", total)
    print("Unique:", unique)
    print("Avg Length:", avg)
```

```python
#task 
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 words:", counter.most_common(10))
print("Least 5 words:", counter.most_common()[-5:])
```

```python
#task 6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
